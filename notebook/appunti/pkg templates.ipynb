## template
set di configurazione per generare pacchetti.  
opzioni:

user::AbstractString="username" <- nome autore
authors::Union{AbstractString, Vector{<:AbstractString}}="name <email> and contributors" <- lista contributori

dir::AbstractString="~/.julia/dev" <- cartella output dove verrà creato il pacchetto
host::AbstractString="github.com": <- URL al servizio di cloud su cui caricare il codice
julia::VersionNumber=v"1.0.0" <- versione minima julia richiesta

plugins::Vector{<:Plugin}=Plugin[]: <- lista plugin
    
si può generare template interattivamente settando opzione interactive = true o usando funzione generate
PkgTemplates.generate("pkgname")
    
(non funziona attraverso notbeook jupyter)

## plugin
i plugin automatizzano operazioni sul pacchetto  
plugin standard:  
+ ProjectFile(; version=v"0.1.0") <- crea project.toml
+ SrcDir(; file="~/work/PkgTemplates.jl/PkgTemplates.jl/templates/src/module.jl") <- crea module entrypoint
+ Tests(; file="~/work/PkgTemplates.jl/PkgTemplates.jl/templates/test/runtests.jl", project=false) <- setup dei test
+ Readme(; file="~/work/PkgTemplates.jl/PkgTemplates.jl/templates/README.md", destination="README.md", inline_badges=false,) <- setup readme
+ License(; name="MIT", path=nothing, destination="LICENSE") <- crea file licenza
+ Git(;
    ignore=String[],
    name=nothing,
    email=nothing,
    branch=nothing,
    ssh=false,
    jl=true,
    manifest=false,
    gpgsign=false,) <- crea repo git
