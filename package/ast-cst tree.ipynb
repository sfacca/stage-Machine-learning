```julia
function parse_file(path::AbstractString)
    code = read(path, String)
    
    Meta.parse("begin $(map(remove_issue,code)) end")
end

function remove_issue(a)    
    if a == '_'
        'T'
    else
        if a == '$'
            'X'
        else
            a
        end
    end
end
```
