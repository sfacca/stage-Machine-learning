```julia
function parse_file(path::AbstractString) 
    temp = fix_file(path)
    res = SemanticModels.ExprModels.Parsers.parsefile(temp)
    rm(temp)
    res
end
```

```julia
function fix_file(path)
    write(string(path,"_fixed.jl"),map(remove_issue,read(path, String)))
    string(path,"_fixed.jl")
end
```
